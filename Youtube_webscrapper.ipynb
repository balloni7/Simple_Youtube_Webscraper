## Driver Setup

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
#Youtube channel you want to scrape
youtube_link = "https://www.youtube.com/@Drututt1/videos"
```

```python
#Go to thaat youtube page
driver.get(youtube_link)
```

## Get All the Videos Titles URLs

```python
## ja funcionou melhor
for _ in range(50):
    driver.find_element(By.TAG_NAME,"body").send_keys(Keys.END)

    # Wait for a short time to allow content to load
    time.sleep(3)  # Adjust the time as needed

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
videos = soup.find_all("div", {"id": "dismissible"})
```

```python
master_list = []

for video in videos:
    data_dict = {}
    
    data_dict["title"] = video.find("a", {"id":"video-title-link"}).text
    data_dict["video_url"] = f"https://www.youtube.com{video.find('a', {'id':'video-title-link'})['href']}"
    
    master_list.append(data_dict)
```
